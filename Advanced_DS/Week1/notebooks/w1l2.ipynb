## Introduction to Linear Regression

This lesson introduces linear regression as a machine learning task. Regression, or the creation of a predictive model from data, is one of the key machine learning tasks. By using linear regression, which can often be solved analytically, we will introduce the concept of minimizing a cost function to determine the optimal model parameters in a straight-forward manner.

### Objectives

By the end of this lesson, you will be able to

- explain the role of the loss function in machine learning,
- articulate how to compute a regression on multiple variables,
- explain the different statistical measures that quantify the quality of a regression, 
- utilize categorical variables in a machine learning model, and
- compute a linear regression mode by using the scikit_learn and statsmodel libraries.

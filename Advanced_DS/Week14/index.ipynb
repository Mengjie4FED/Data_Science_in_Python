# Week 14: Introduction to Probabilistic Modeling

This module introduces probabilistic modeling, and in particular Bayesian Model fitting. In the Bayesian approach, we compute the likelihood of a model given observed data. This is a popular technique since we obtain posterior distributions on our model parameters. First, explore the general concepts in Bayesian model fitting. This will require using the PyMC3 library to compute model parameters, sample from the likelihood, and evaluate the resulting fit. Next, we will learn about hierarchical model fitting, where both the entire data and subsets of the data are fit at the same time. Finally, we will explore general linear modeling, where he model errors are allowed to be non-Gaussian.


## Learning Objectives

#### By the end of this module, you should be able to:

- explain Bayesian model regression,
- describe the fundamentals of hierarchical model fitting,
- explain the basic principles of general linear modeling, and
- use the PyMC3 library to perform Bayesian model fitting, including both hierarchical models and general linear models.
